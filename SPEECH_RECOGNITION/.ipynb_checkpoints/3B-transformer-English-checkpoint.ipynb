```python
from transformers import Wav2Vec2Processor, Wav2Vec2ForCTC
import soundfile as sf
import torch
from time import time
from tqdm import tqdm
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim


# load model and tokenizer
processor = Wav2Vec2Processor.from_pretrained("facebook/wav2vec2-large-960h")
model = Wav2Vec2ForCTC.from_pretrained("facebook/wav2vec2-large-960h").eval().to('cuda')
# load dummy dataset and read soundfiles
```

```python
minimum = 9999999
with torch.no_grad():
  for nb_splits in range(30,100,5):
    t0 =time()
    ds = sf.read('gdrive/MyDrive/Raw-Video-Barack-Obama_s-2008-acceptance-speech-_LEo7lzfpdCU_.flac')
    input_values = processor(ds[0], return_tensors="pt", padding="longest",num_process=6,sampling_rate=16000).input_values  # Batch size 1
    # retrieve
    operations = input_values.shape[1] // nb_splits
    liste = []
    for i in tqdm(range(nb_splits)):
      output =  model(input_values[:,i*operations:(i+1)*operations].to(dev))
      liste.append(torch.argmax(output.logits, dim=-1))
    liste2=[]
    for output in tqdm(liste):
      liste2.append(output.detach().to(device='cpu', dtype=torch.long))
    for output in liste2:
      processor.batch_decode(output)
    delta_time = time()-t0
    print(delta_time)
    if delta_time < minimum:
      minimum = delta_time
      value = nb_splits
      print('value:',value,'delta time',delta_time)
```
